# Crosswalk

```python
import openpyxl
from openpyxl import Workbook, load_workbook
```

```python
# append labels to a data dictionary
def append_labels(algo, data):
    for row in range(2,1000):
        if algo.cell(row,1).value is None:
            break
        data[algo.cell(row,1).value] = {}
        for col in range(2,11):
            cell_value_class = algo.cell(1,col).value
            cell_value_id = algo.cell(row,col).value
            data[algo.cell(row,1).value].update({cell_value_class:cell_value_id})
```

## Hashmap

```python
# John
hashmap = load_workbook('../labels/john/Sort-zOthers.xlsx')

hashmapWS = hashmap.active

john = {}

append_labels(hashmapWS,john)

print(john)

counter = 0
for x in john:
    counter += 1
print("\nData count:",counter) 

#Gerd
hashmap = load_workbook('../labels/gerd/Sort-zOthers.xlsx')

hashmapWS = hashmap.active

gerd = {}

append_labels(hashmapWS,gerd)

print(gerd)

counter = 0
for x in gerd:
    counter += 1
print("\nData count:",counter) 

#Karl
hashmap = load_workbook('../labels/karl/Sort-zOthers.xlsx')

hashmapWS = hashmap.active

karl = {}

append_labels(hashmapWS,karl)

print(karl)

counter = 0
for x in karl:
    counter += 1
print("\nData count:",counter) 
```

## Create CrosswalkWorksheet

```python
wb = Workbook()
ws = wb.active

headings = ['Date', 'Filename', 'Gerd', 'John', 'Karl', 'FINAL']
ws.append(headings)

wb.save("Sort-zOthers.xlsx")
```

## Input individual labels per file, Check if similar labels, change color of rows with different labels to red for easy finding

```python
from openpyxl.utils import get_column_letter
from openpyxl.styles import Font
import datetime

date = datetime.datetime.now()

row = 2

for file in john:
    jlabels = []
    glabels = []
    klabels = []
    for (algo, label) in john[file].items():
        if john[file][algo] == 1:
            jlabels.append(algo)
            
    for (algo, label) in gerd[file].items():
        if gerd[file][algo] == 1:
            glabels.append(algo)
    
    try:
        for (algo, label) in karl[file].items():
            if karl[file][algo] == 1:
                klabels.append(algo)
    except KeyError:
        print(file)

    ws["B" + str(row)] = file
    
    for xlabel in glabels:
        if ws["C" + str(row)].value is not None and ws["C" + str(row)].value != xlabel:
            ws["C" + str(row)] = str(ws["C" + str(row)].value) + str(xlabel) + ";"
        else:
            ws["C" + str(row)] = str(xlabel) + ";"
    
    for xlabel in jlabels:
        if ws["D" + str(row)].value is not None and ws["D" + str(row)].value != xlabel:
            ws["D" + str(row)] = str(ws["D" + str(row)].value) + str(xlabel) + ";"
        else:
            ws["D" + str(row)] = str(xlabel) + ";"
    
    for xlabel in klabels:
        if ws["E" + str(row)].value is not None and ws["E" + str(row)].value != xlabel:
            ws["E" + str(row)] = str(ws["E" + str(row)].value) + str(xlabel) + ";"
        else:
            ws["E" + str(row)] = str(xlabel) + ";"
    
#     ws["C" + str(row)] = str(glabels)
#     ws["D" + str(row)] = str(jlabels)
#     ws["E" + str(row)] = str(klabels)
    
    # set final labels for rows with same labels among the researchers
#     if glabels == jlabels and glabels == klabels and jlabels == klabels:
#         ws["A" + str(row)] = date.strftime("%x")
# #         ws["F" + str(row)] = str(glabels)
#         for xlabel in glabels:
#             if ws["F" + str(row)].value is not None and ws["F" + str(row)].value != xlabel:
#                 ws["F" + str(row)] = str(ws["F" + str(row)].value) + str(xlabel) + ";"
#             else:
#                 ws["F" + str(row)] = str(xlabel) + ";"

    if ws["C" + str(row)].value == ws["D" + str(row)].value and ws["C" + str(row)].value == ws["E" + str(row)].value and ws["D" + str(row)].value == ws["E" + str(row)].value:
        ws["A" + str(row)] = date.strftime("%x")
        ws["F" + str(row)] = ws["C" + str(row)].value
        
    
    else:
        for col in range(1, len(headings) + 1):
            ws[get_column_letter(col) + str(row)].font = Font(color = "FF0000")
        
    row = row + 1

wb.save("Sort-zOthers.xlsx")
            
```
